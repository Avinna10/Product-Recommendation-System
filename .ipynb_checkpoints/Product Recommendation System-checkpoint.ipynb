<h3>Importing Required Libraries</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import mean_absolute_error
from sklearn.metrics.pairwise import cosine_similarity
```

<h3>Reading a JSON dataset file</h3>

```python
# r represents a raw string
product = r'dataset\products.json'
rating = r'dataset\ratings.json'

df_product = pd.read_json(product)
df_rating = pd.read_json(rating)
```

```python
df_product.head()
```

```python
df_rating.head()
```

<h3>Data Pre-Processing</h3>

<span>Removing duplicate rows from the product dataframe</span>

```python
# dropping product rows having same ID and displaying 5 records
df_product.head().drop_duplicates('ProductID')
```

```python
# checking the NAN records in rating column
df_rating = df_rating[df_rating['rating'].notna()]
df_rating.head()
```

<h3>Merging rating and product datasets</h3>

```python
combine_product_rating=pd.merge(df_rating, df_product,on="ProductID")
combine_product_rating.head()
```

<h3>Data Train Test Split</h3>

```python
# specifing three columns to split the dataset
product_users = combine_product_rating[['user_id', 'ProductID', 'rating']]

# split 80% of data as train set and 20% as test set
X_train, X_test, y_train, y_test = train_test_split(product_users, product_users['rating'].values, test_size=0.2)
```

<h3>Model Fit</h3>

```python
model_knn = KNeighborsClassifier(n_neighbors=3, metric='cosine')
model_knn.fit(X_train, y_train)

# make predictions on test data
prediction = model_knn.predict(X_test)

# calculate accuracy based on data  predicted
accuracy = accuracy_score(y_test, prediction)
print("Accuracy of the model: ", accuracy)
```

<h3>Data Pivot</h3>

```python
#data pivot is created. NaN value is not filled with other values because NaN check is used in recommendation logic
train_data = X_train.pivot(index='user_id', columns='ProductID', values='rating')
train_data.head()
```

<h3>Cosine Similarity</h3>

```python
# generating the similarity value between the users by filling 0 in NAN value
user_similarity_cosine = cosine_similarity(train_data.fillna(0))

#converting similarity array into data frame by assigning the index and column as id of user
user_similarity = pd.DataFrame(user_similarity_cosine, columns=train_data.index, index=train_data.index)
user_similarity.head()
```

<h3>Model Tuning</h3>

```python
# instance of the KNeighborsClassifier using cosine metric
knn = KNeighborsClassifier(metric='cosine')

#maintains the proportion of class in every split
cv = ShuffleSplit(n_splits=2, random_state=42)

# Specifing the hyperparameters and possible values to tune
param_grid = {'n_neighbors': np.arange(1, 25), 'weights':['uniform','distance']}

# instance of the GridSearchCV
grid = GridSearchCV(knn, param_grid, cv=cv)

# fit training set to grid
grid.fit(X_train, y_train)

# generates the best parameter
print("Best parameters:", grid.best_params_)
# generates the best score
print("Best score:", grid.best_score_)
```

<h3>Accuracy Calculation</h3>

```python
# make predictions on test data
tuned_prediction = grid.predict(X_test)

# calculate accuracy based on data  predicted
tuned_accuracy = accuracy_score(y_test, tuned_prediction)
print("Accuracy of the model: ", tuned_accuracy)
```

<h3>Performance Metrics</h3>

```python
# calculation of MAE
test_mae = mean_absolute_error(y_test, tuned_prediction)
print("MAE: ", test_mae)

# calculation of RMSE 
test_rmse = np.sqrt(np.mean((tuned_prediction - prediction) ** 2))
print("RMSE: ", test_rmse)
```

<h3>Confusion Matrix</h3>

```python
# generating confusion matrix
matrix = confusion_matrix(y_test, tuned_prediction)
print(matrix)

# plotting confusion matrix in graph
print("\n\t---Graphical representation---")
matrix = ConfusionMatrixDisplay(matrix)
matrix.plot()
plt.show()
```

<h3>Generating recommendations</h3>

```python
# Select a random user from the dataset
random_user = train_data.sample()
user_id = random_user.index.values[0]

# Remove the selected user ID from the similarity
user_similarity.drop(index=user_id, inplace=True)
user_similarity.head()
```

```python
# Get top 5 similar users having similarity score above 0.4
similar_users = user_similarity[user_similarity[user_id]>0.4][user_id].sort_values(ascending=False)[:5]

# Print out top 5 similar users
print(f'The similar users for user {user_id} are\n', similar_users)
```

```python
# sets the products, user selected has rated
user_rated_product = train_data[train_data.index == user_id].dropna(axis=1, how='all')

# stores the products that similar user has rated and removes the unrated products
similar_products = train_data[train_data.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_products
```

```python
# removes the product that the selected user has already rated 
similar_products.drop(user_rated_product.columns,axis=1, inplace=True, errors='ignore')
similar_products
```

```python
# stores the product scores in dictionary
product_score = {}

for i in similar_products.columns:
  # Get the ratings for product i
  product_rating = similar_products[i]
  # variable to store the score
  total_score = 0
  # variable to store the number of scores
  count_score = 0
    
  # Loop through similar users
  for j in similar_users.index:
    # if the product has been rated
    if pd.isna(product_rating[j]) == False:
      # Score is the sum of user similarity score multiply by the product rating
      score = similar_users[j] * product_rating[j]
      # Add the score to the total score for the product so far
      total_score += score
      # Add 1 to the count_score
      count_score +=1
  # Get the average score for the item
  product_score[i] = total_score / count_score

# Convert dictionary to pandas dataframe
product_score = pd.DataFrame(product_score.items(), columns=['ProductID', 'product_score'])

# Sort the products by score
ranked_product_score = product_score.sort_values(by='product_score', ascending=False)
ranked_product_score.head()
```

```python
# merging the ranked_product_score with product dataset
merged_df = pd.merge(ranked_product_score, df_product, on='ProductID')

# gets the name of the product
def get_name(id):
    name = merged_df.loc[merged_df['ProductID'] == id, 'Name'].values[0]
    return name

# list to store name of recommended items
name = []

for i in ranked_product_score['ProductID']:
    name.append(get_name(i))
    
print(f'The top recommendations for user {user_id} based on similar users are:\n')
for i in range(5):
    print(name[i])
```
